```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g1(a):
    return 1./(1.+np.exp(-a))
def g2(a):
    return a

N = 201
inputs = np.linspace(-2, 2, N)
w1 = np.random.randn(100,1)
w2 = np.random.randn(50,100)
w3 = np.random.randn(1,50)
```

```python
h1 = g1(np.multiply(w1,inputs))
print(h1.shape)
h2 = g1(np.dot(w2,h1))
outputs = np.dot(w3,h2)
outputs.shape
```

```python
plt.plot(inputs,outputs.T)
```

removing the non-linearity sigmoid function

```python
h1 = g2(np.multiply(w1,inputs))
h2 = g2(np.dot(w2,h1))
outputs = g2(np.dot(w3,h2))
```

```python
plt.plot(inputs,outputs.T)
```

scale w

```python
w1_n = w1 * 0.1
w2_n = w2 * 0.1
w3_n = w3 * 0.1
h1 = g1(np.multiply(w1_n,inputs))
h2 = g1(np.dot(w2_n,h1))
outputs = g2(np.dot(w3_n,h2))
```

```python
plt.plot(inputs,outputs.T)
```
